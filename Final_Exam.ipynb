# **ไฟล์ข้อมูล alldb**

```python
#โหลดข้อมูลของไฟล์ alldb เข้ามาเพื่อนแสดงค่าข้อมูลต่างๆ
```

```python
alldb = pd.read_csv('alldb.csv') #ใช้คำสั่ง .read_csv เพื่ออ่านไฟล์ที่ใช้นามสกุล csv
print(alldb.shape)
alldb
```

```python
# ลบ คอลัมน์ที่ไม่ต้องการออก โดยใช้คำสั่ง .drop
```

```python
alldb = alldb.drop(columns=['Unnamed: 0','care','pid'])     # คอลัมน์ที่โดนลบออกไปคือ  Unnamed: 0 , care , pid
print(alldb.shape)
alldb
```

```python
alldb.isnull().sum() # ใช้คำสั่ง Sum  เพื่อดูว่าแต่ละ Row มีค่าว่างหรือไม่
```

# **แปลงค่าของข้อมูล user**

```python
# กำหนดและแปลงค่าของ Columns user ให้เป็นตัวเลข 5 หลัก (ใช้เฉพาะตัวเลขเท่านั้น)
# ใช้คำสั่ง .str.isdigit เพื่อตรวจสอบว่าอักขระทั้งหมดเป็นตัวเลขหรือไม่
# อ้างอิงที่มาโค้ด https://www.w3schools.com/python/ref_string_isdigit.asp
```

```python
alldb = alldb[(alldb['user'].astype(str).str.isdigit()) & (alldb['user'].str.len() == 5)]  # เลือกมาเฉพาะ คอลัมน์ user ที่เป็นตัวเลข และ ที่มีตัวเลข เท่ากับ 5 ตัว
print(alldb.shape)                                                                             
alldb
```

```python
alldb.isnull().sum()  # ใช้คำสั่งอีกครั้ง Sum  เพื่อดูว่าแต่ละ Columns มีค่าว่างหรือไม่
```
